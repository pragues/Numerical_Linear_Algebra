### Representations: 
$A \in \mathbb{R}^{m \times n}$, 列用小写 `a_i`， $a_i$。 行的表示感觉很混乱，$a_{(1)}^T$ 理解的话：把$A$一行一行的切，每行拿出来是一个"$m \times 1$"的向量$a_{(1)}$，然后向量是竖着的，再放回去成$a_{(1)}^T$。 古早MAT2041线代课上貌似也有$A_{(1)}$来表示行。上标写的话一般是第几次iteration. 

算法里写的时候$A_{[i, j]}$表示$i$行$j$列的元素, $A_{[i, :]}$表示单拎出来第$i$行，$A_{[:, j]}$表示单拎出来第$j$列，dummy variables随便,`:`符号可以理解为"all"。

$ 
A = \begin{bmatrix}
   a_1 & a_2 & ... & a_n
\end{bmatrix} 
= \begin{bmatrix}
   A_{(1)} \\
   A_{(2)}  \\
   ... \\
   A_{(m)}  \\
   \end{bmatrix}
= \begin{bmatrix}
   a_{(1)}^T \\
   a_{(2)}^T  \\
   ... \\
   a_{(m)}^T \\
   \end{bmatrix}
$ 


#### 左乘与右乘. Left-rows, right-columns

看不出来的时候通过矩阵分块在脑子里想象一下

**Left-rows**: $L \in \mathbb{R}^{k \times m}, A \in \mathbb{R}^{m \times n}$ 左乘结果的每一行可以看做$A$ rows 的线性组合

$C=L \cdot A = \begin{bmatrix}
   L^{(1)} A \\
   L^{(2)} A \\
   ... \\
   L^{(k)} A \\
\end{bmatrix} $

So each row we have $L^{(i)} A = 
\begin{bmatrix}
   L_{i1} & L_{i2} & ... & L_{im}
\end{bmatrix}  \begin{bmatrix}
   A^{(1)} \\
   A^{(2)}  \\
   ... \\
   A^{(m)}  \\
   \end{bmatrix} = \sum_{j=1}^{m} L_{ij} \cdot A^{(j)}$, as a linear combination of the rows of A

<br>

<br>

**Right-columns**: $A \in \mathbb{R}^{m \times n}, R \in \mathbb{R}^{n \times k}$ 右乘结果的每一列可以看做$A$ columns的线性组合

$C=A \cdot R = \begin{bmatrix}
   A r_1 & A r_2 & ... & Ar_k
\end{bmatrix} 
$

For column i of $C$, we have:

$c_i = A r_i = \begin{bmatrix}
   a_1 & a_2 & ... & a_n
\end{bmatrix}\begin{bmatrix}
   r_{1i} \\
   r_{2i}  \\
   ...\\
   r_{ni}  \\
   \end{bmatrix} = \sum_{j=1}^{n} a_j r_{ji} $, as a linear combination of the columns of $A$
