### Representations: 
$A \in \mathbb{R}^{m \times n}$

$ 
A = \begin{bmatrix}
   a_1 & a_2 & ... & a_n
\end{bmatrix} 
= \begin{bmatrix}
   A^{(1)} \\
   A^{(2)}  \\
   ... \\
   A^{(m)}  \\
   \end{bmatrix}
= \begin{bmatrix}
   a_{(1)}^T \\
   a_{(2)}^T  \\
   ... \\
   a_{(m)}^T \\
   \end{bmatrix}
$ 


#### 左乘与右乘. Left-rows, right-columns

每次记不住通过矩阵分块在脑子里想象一下

Left-rows: $L \in \mathbb{R}^{k \times m}, A \in \mathbb{R}^{m \times n}$

$C=L \cdot A = \begin{bmatrix}
   L^{(1)} A \\
   L^{(2)} A \\
   ... \\
   L^{(k)} A \\
\end{bmatrix} $

So each row we have $L^{(i)} A = 
\begin{bmatrix}
   L_{i1} & L_{i2} & ... & L_{im}
\end{bmatrix}  \begin{bmatrix}
   A^{(1)} \\
   A^{(2)}  \\
   ... \\
   A^{(m)}  \\
   \end{bmatrix} = \sum_{j=1}^{m} L_{ij} \cdot A^{(j)}$, as a linear combination of the rows of A

<br>

Right-columns: $A \in \mathbb{R}^{m \times n}, R \in \mathbb{R}^{n \times k}$

$C=A \cdot R = \begin{bmatrix}
   A r_1 & A r_2 & ... & Ar_k
\end{bmatrix} 
$

For column i of $C$, we have:

$c_i = A r_i = \begin{bmatrix}
   a_1 & a_2 & ... & a_n
\end{bmatrix}\begin{bmatrix}
   r_{1i} \\
   r_{2i}  \\
   ...\\
   r_{ni}  \\
   \end{bmatrix} = \sum_{j=1}^{n} a_j r_{ji} $, as a linear combination of the columns of $A$
