gmre from scratch:

arnoldi+ least square method

```python
import numpy as np

def gmres_simple(A, b, k):
    n = len(b)
    Q = np.zeros((n, k+1))
    H = np.zeros((k+1, k))

    beta = np.linalg.norm(b)
    Q[:, 0] = b / beta

    for j in range(k):
        v = A @ Q[:, j]
        for i in range(j+1):
            H[i, j] = np.dot(Q[:, i], v)
            v -= H[i, j] * Q[:, i]

        H[j+1, j] = np.linalg.norm(v)
        if H[j+1, j] != 0:
            Q[:, j+1] = v / H[j+1, j]

    # solve min ||β e1 – H y||
    y, *_ = np.linalg.lstsq(H[:k+1, :k], beta * np.eye(k+1)[0], rcond=None)
    x = Q[:, :k] @ y
    return x

# Test
A = np.array([[4., 1.], [1., 3.]])
b = np.array([1., 2.])

print(gmres_simple(A, b, 2))
```
